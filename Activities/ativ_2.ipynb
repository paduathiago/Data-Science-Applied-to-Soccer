```python
BRA1 = pd.read_csv('../data/brazilian_league_20222.csv', sep=',')
BRA1.drop(columns=['Match Report', 'Notes', 'Attendance', 'Venue', 'Time', 'Referee', 'Day'], inplace=True)
BRA1 = BRA1.dropna()
BRA1 = BRA1.rename(columns={'xG':'xg_pro', 'xG.1':'xg_con'})
display(BRA1)
```

```python
def assign_xg(match, home=True):
    if home:
        xg_pro = match['xg_pro']
        xg_con = match['xg_con']
    else:
        xg_pro = match['xg_con']
        xg_con = match['xg_pro']
    return xg_pro, xg_con, xg_pro - xg_con
```

```python
def rolling_average(teams, window):
    teams_dataframe = {}
    for team in teams:
        team_df = pd.DataFrame(columns=['xg_pro', 'xg_con', 'xg_diff', 'Game'])
        matches = BRA1.loc[(BRA1['Home'] == team) | (BRA1['Away'] == team)]
        game = 0
        for i, match in matches.iterrows():
            game += 1
            if match['Home'] == team:
                xg_pro, xg_con, xg_diff = assign_xg(match)
            else:
                xg_pro, xg_con, xg_diff = assign_xg(match, home=False)
            
            team_df.at[i, "xg_pro"] = xg_pro
            team_df.at[i, "xg_con"] = xg_con
            team_df.at[i, "xg_diff"] = xg_diff
            team_df.at[i, "Game"] = game

        team_df['xg_pro_RA'] = team_df['xg_pro'].rolling(window=window, win_type='triang', min_periods=1).mean()
        team_df['xg_con_RA'] = team_df['xg_con'].rolling(window=window, win_type='triang', min_periods=1).mean()
        team_df['xg_diff_RA'] = team_df['xg_diff'].rolling(window=window, win_type='triang', min_periods=1).mean()

        teams_dataframe[team] = team_df
    
    return teams_dataframe
```

```python
def get_mean_statistic(statistic: str, league_dataframe: pd.DataFrame):
    rounds_df = pd.DataFrame(columns=['statistical_value'])
    new_statistic = {}
    for i in range(1, 39):
        if statistic != 'xg_diff':
            new_statistic[i] = league_dataframe.loc[(league_dataframe['Wk'] == i), statistic]
        else:
            xg_pro = league_dataframe.loc[(league_dataframe['Wk'] == i), 'xg_pro']
            xg_con = league_dataframe.loc[(league_dataframe['Wk'] == i), 'xg_con']
            new_statistic[i] = xg_pro - xg_con
        rounds_df.at[i, 'statistical_value'] = new_statistic[i].mean()

    return rounds_df
```

```python
def plot_RA(league, team_dfs, statistic, teams_to_plot, window):
    fig, ax = plt.subplots(figsize=(8, 4))
    
    df_mean_statistic = get_mean_statistic(statistic, league)
    ax.plot(df_mean_statistic.index, df_mean_statistic['statistical_value'].rolling(window=window, win_type='triang', min_periods=1).mean(),
             linewidth=3, linestyle='dashed',color='black', alpha = 1, zorder = 2, label = 'Média Geral')

    colors = ['green', 'red', 'blue', 'purple']
    alphas = [1, 1, 1, 1]

    for club, color, alpha in zip(teams_to_plot, colors, alphas):
        ax.plot(team_dfs[club]['Game'],  team_dfs[club][f'{statistic}_RA'], linewidth=2, linestyle='-',color=color, alpha = alpha, zorder = 2, label = club)

    ax.set_title(f"{statistic} {window} rounds Moving Average  - Brazilian League 2022", fontsize = 18, pad=20)  # set title
    ax.set_xticklabels([])  # no text
    ax.set_xticks(np.arange(1, 39))
    ax.tick_params(axis='x', color='white', which='both', labelsize = 20)  # disable black ticks, but big text xaxis
    ax.tick_params(axis='y', color='grey', labelsize = 14)  # grey y ticks
    ax.grid(zorder = 1, linestyle = "--", alpha = 0.5)  # add grid
    
    # disable spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    #add axis labels
    ax.set_ylabel(f"{statistic} Moving Average per game", fontsize=16, labelpad=10)
    ax.set_xlabel('Round', fontsize=18, labelpad=10)
    plt.legend(bbox_to_anchor=(1.0, 0.5), loc="center left", fontsize=12, edgecolor='white')  # make legend
    plt.show()
```

```python
def RA_calc_and_plot(league: pd.DataFrame, teams_to_plot: list, window: int=5):
    statistics = ['xg_pro', 'xg_con', 'xg_diff']
    teams = league['Home'].unique()
    teams_df = rolling_average(teams, window=window)
    for statistic in statistics:
        plot_RA(league, teams_df, statistic, teams_to_plot, window)
```

```python
def color_generator(number_of_colors):
    # generates a list of random colors
    colors = []
    for _ in range(number_of_colors):
        r = random.random()
        g = random.random()
        b = random.random()
        colors.append((r, g, b))

    return colors
```

```python
def plot_radar(params, players, low, high, lower_is_better):
        radar = Radar(params, low, high,
              lower_is_better=lower_is_better,
              # whether to round any of the labels to integers instead of decimal places
              round_int=[False]*len(params),
              num_rings=4,  # the number of concentric circles (excluding center circle)
              # if the ring_width is more than the center_circle_radius then
              # the center circle radius will be wider than the width of the concentric circles
              ring_width=1, center_circle_radius=1)

        colors = color_generator(len(players))

        fig, ax = radar.setup_axis(figsize=(10,10))
        rings_inner = radar.draw_circles(ax=ax, facecolor='#D3D3D3', edgecolor='#fc5f5f')
        
        # add Title
        fig.text(
                0.515, 0.942,
                f"Players per 90:\n Percentile Rank vs Brazilian League {players[0].pos} | Season 2022",
                size=18,
                ha="center",fontproperties=font_bold.prop, color="#000000"
        )

        for i, player in enumerate(players):
                player_values = player.get_data()
                player_name = player.get_name()
                radar1, vertices1 = radar.draw_radar_solid(player_values, ax=ax,
                                                        kwargs={'facecolor': colors[i],
                                                                'alpha': 0.5,
                                                                'edgecolor': '#216352',
                                                                'lw': 3})

                ax.scatter(vertices1[:, 0], vertices1[:, 1],
                        c=colors[i], edgecolors='#502a54', marker='o', s=150, zorder=2, label=player_name) 
        plt.legend(bbox_to_anchor=(1.0, 0.5), loc="center left", fontsize=12, edgecolor='white')  # make legend

        range_labels = radar.draw_range_labels(ax=ax, fontsize=25, fontproperties=robotto_thin.prop)
        param_labels = radar.draw_param_labels(ax=ax, fontsize=25, fontproperties=robotto_thin.prop)
```

```python
def summary_per_90(dataframe, params):
    df = pd.DataFrame()
    df['Player'] = dataframe['Player']
    for param in params:
        df[param] = dataframe.apply(lambda x: x[param]/x["90s"], axis=1)
    return df
```

```python
class Player:
    def __init__(self, name, pos) -> None:
        self.name = name
        self.pos = pos
        self.params = []
        if self.pos == 'fw':
            aux_shooting_fw = shooting[shooting['Pos'].apply(lambda x:'FW' in x)].dropna()
            shooting_fw = summary_per_90(aux_shooting_fw, ['Gls', 'npxG', 'SoT'])

            aux_passing_fw = passing[passing['Pos'].apply(lambda x:'FW' in x)].dropna()
            passing_fw = summary_per_90(aux_passing_fw, ['Ast', '1/3'])
            
            aux_possession_fw = possession[possession['Pos'].apply(lambda x:'FW' in x)].dropna()
            possession_fw = summary_per_90(aux_possession_fw, ['Att Pen'])

            self.params.append(stats.percentileofscore(shooting_fw['Gls'], float(shooting_fw.loc[(shooting_fw['Player'] == self.name), 'Gls'])))
            self.params.append(stats.percentileofscore(shooting_fw['npxG'], float(shooting_fw.loc[(shooting_fw['Player'] == self.name), 'npxG'])))
            self.params.append(stats.percentileofscore(shooting_fw['SoT'], float(shooting_fw.loc[(shooting_fw['Player'] == self.name), 'SoT'])))
            self.params.append(stats.percentileofscore(passing_fw['Ast'], float(passing_fw.loc[(passing_fw['Player'] == self.name), 'Ast'])))
            self.params.append(stats.percentileofscore(passing_fw['1/3'],float(passing_fw.loc[(passing_fw['Player'] == self.name), '1/3'])))
            self.params.append(stats.percentileofscore(possession_fw['Att Pen'], float(possession_fw.loc[(possession['Player'] == self.name), 'Att Pen'])))
        elif self.pos == 'mf':
            aux_shooting_mf = shooting[shooting['Pos'].apply(lambda x:'MF' in x)].dropna()
            shooting_mf = summary_per_90(aux_shooting_mf, ['Gls'])

            aux_passing_mf = passing[passing['Pos'].apply(lambda x:'MF' in x)].dropna()
            passing_mf = summary_per_90(aux_passing_mf, ['Ast', '1/3'])

            aux_possession_mf = possession[possession['Pos'].apply(lambda x:'MF' in x)].dropna()
            possession_mf = summary_per_90(aux_possession_mf, ['Succ'])

            aux_miscellaneous_mf = miscellaneous[miscellaneous['Pos'].apply(lambda x:'MF' in x)].dropna()
            miscellaneous_mf = summary_per_90(aux_miscellaneous_mf, ['Recov'])

            aux_defensive_mf = defensive[defensive['Pos'].apply(lambda x:'MF' in x)].dropna()
            defensive_mf = summary_per_90(aux_defensive_mf, ['Inte'])

            self.params.append(stats.percentileofscore(shooting_mf['Gls'], float(shooting_mf.loc[(shooting_mf['Player'] == self.name), 'Gls'])))
            self.params.append(stats.percentileofscore(passing_mf['Ast'], float(passing_mf.loc[(passing_mf['Player'] == self.name), 'Ast'])))
            self.params.append(stats.percentileofscore(passing_mf['1/3'], float(passing_mf.loc[(passing_mf['Player'] == self.name), '1/3'])))
            self.params.append(stats.percentileofscore(possession_mf['Succ'], float(possession_mf.loc[(possession_mf['Player'] == self.name), 'Succ'])))
            self.params.append(stats.percentileofscore(miscellaneous_mf['Recov'], float(miscellaneous_mf.loc[(miscellaneous_mf['Player'] == self.name), 'Recov'])))
            self.params.append(stats.percentileofscore(defensive_mf['Inte'], float(defensive_mf.loc[(defensive_mf['Player'] == self.name), 'Inte'])))
        elif self.pos == 'df':
            aux_miscellaneous_df = miscellaneous[miscellaneous['Pos'].apply(lambda x:'DF' in x)].dropna()
            miscellaneous_df = summary_per_90(aux_miscellaneous_df, ['Won', 'Recov', 'Fls'])

            aux_defensive_df = defensive[defensive['Pos'].apply(lambda x:'DF' in x)].dropna()
            defensive_df = summary_per_90(aux_defensive_df, ['Pass', 'Inte', 'Tkl'])
            
            self.params.append(stats.percentileofscore(defensive_df['Pass'], float(defensive_df.loc[(defensive_df['Player'] == self.name), 'Pass'])))
            self.params.append(stats.percentileofscore(defensive_df['Inte'], float(defensive_df.loc[(defensive_df['Player'] == self.name), 'Inte'])))
            self.params.append(stats.percentileofscore(defensive_df['Tkl'], float(defensive_df.loc[(defensive_df['Player'] == self.name), 'Tkl'])))
            self.params.append(stats.percentileofscore(miscellaneous_df['Won'], float(miscellaneous_df.loc[(miscellaneous_df['Player'] == self.name), 'Won'])))
            self.params.append(stats.percentileofscore(miscellaneous_df['Recov'], float(miscellaneous_df.loc[(miscellaneous_df['Player'] == self.name), 'Recov'])))
            self.params.append(stats.percentileofscore(miscellaneous_df['Fls'], float(miscellaneous_df.loc[(miscellaneous_df['Player'] == self.name), 'Fls'])))
            
    def get_data(self):
        return self.params
    
    def get_name(self):
        return self.name
```

```python
def plot_comparative_radar(params: list, pos: str, player_names: list, lower_is_better: list=[]):
    players = []
    for name in player_names:
        player = Player(name, pos)
        players.append(player)

    low = np.zeros(len(params))
    high = np.full(len(params), 100)
    
    plot_radar(params, players, low, high, lower_is_better)
```
